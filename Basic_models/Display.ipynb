```python
import random
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
# шлях до справжніх даних
real_path = ""
# шлях до синтезованих даних
fake_path = ""

real_files = os.listdir(real_path)
fake_files = os.listdir(fake_path)

real_file = random.choice(real_files)
fake_file = random.choice(fake_files)

raudio, sr1 = librosa.load(os.path.join(real_path, real_file), sr=None)
faudio, sr2 = librosa.load(os.path.join(fake_path, fake_file), sr=None)

plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
librosa.display.waveshow(raudio, sr=sr1, color='b')
plt.xlabel('Час')
plt.ylabel('Амплітуда')
plt.title('Waveform для реального аудіофайлу')

plt.subplot(2, 1, 2)
librosa.display.waveshow(faudio, sr=sr2, color='r')
plt.xlabel('Час')
plt.ylabel('Амплітуда')
plt.title('Waveform для фейкового аудіофайлу')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
real_spec = librosa.feature.melspectrogram(y=raudio, sr=sr1)
librosa.display.specshow(librosa.power_to_db(real_spec, ref=np.max), sr=sr1, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f дБ')
plt.title('Мел-спектрограма для реального аудіофайлу')

plt.subplot(2, 1, 2)
fake_spec = librosa.feature.melspectrogram(y=faudio, sr=sr2)
librosa.display.specshow(librosa.power_to_db(fake_spec, ref=np.max), sr=sr2, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f дБ')
plt.title('Мел-спектрограма для фейкового аудіофайлу')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
real_spec = librosa.feature.melspectrogram(y=raudio, sr=sr1)
librosa.display.specshow(real_spec, sr=sr1, x_axis='time', y_axis='mel')

plt.colorbar(label='Amplitude')
plt.title('Мел-спектрограма без нормалізації для реального аудіофайлу')

plt.subplot(2, 1, 2)
fake_spec = librosa.feature.melspectrogram(y=faudio, sr=sr2)
librosa.display.specshow(fake_spec, sr=sr2, x_axis='time', y_axis='mel')

plt.colorbar(label='Amplitude')
plt.title('Мел-спектрограма без нормалізації для фейкового аудіофайлу')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
real_mfcc = librosa.feature.mfcc(y=raudio, sr=sr1, n_mfcc=64) # Число MFCC коефіцієнтів - 13
librosa.display.specshow(real_mfcc, sr=sr1, x_axis='time')
plt.colorbar()
plt.title('MFCC для реального аудіофайлу')

plt.subplot(2, 1, 2)
fake_mfcc = librosa.feature.mfcc(y=faudio, sr=sr2, n_mfcc=13)
librosa.display.specshow(fake_mfcc, sr=sr2, x_axis='time')
plt.colorbar()
plt.title('MFCC для фейкового аудіофайлу')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
real_spec = librosa.stft(y=raudio)
librosa.display.specshow(librosa.amplitude_to_db(np.abs(real_spec), ref=np.max), sr=sr1, x_axis='time', y_axis='linear')
plt.colorbar(format='%+2.0f дБ')
plt.title('Спектрограма для реального аудіофайлу')

plt.subplot(2, 1, 2)
fake_spec = librosa.stft(y=faudio)
librosa.display.specshow(librosa.amplitude_to_db(np.abs(fake_spec), ref=np.max), sr=sr2, x_axis='time', y_axis='linear')
plt.colorbar(format='%+2.0f дБ')
plt.title('Спектрограма для фейкового аудіофайлу')

plt.tight_layout()
plt.show()
```
